```python
import pandas as pd
```

```python
data = pd.read_csv('house_price.csv')
data
```

```python
data1 = data.dropna()#处理缺失值
data2 = pd.get_dummies(data1[['dist','floor']])#返回处理后数据
data2
```

```python
data3 = data2.drop(['dist_shijingshan','floor_high'],axis=1)
# pd.set_option('display.max_rows', None)   #显示行
data3
```

```python
data4 = pd.concat([data3,data1[['roomnum','halls','AREA','subway','school','price']]],axis=1)
data4
```

```python
X = data4.iloc[:,:-1]
y = data4.iloc[:,-1:]
from sklearn.model_selection import train_test_split#将矩阵随机划分为训练子集和测试子集
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1,test_size=0.3)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression#调用sklearn中的线性回归
model = LinearRegression().fit(X_train,y_train)
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))#打分
```

```python
model.predict([[1,0,0,0,0,0,1,1,0,46.06,1,0]])#测试数据
```

```python
model.coef_#显示塞塔
```

```python
model.intercept_#截距
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=0.1).fit(X_train,y_train)
print('训练分数:%f'%(ridge.score(X_train,y_train)))
print('测试分数:%f'%(ridge.score(X_test,y_test)))
print(ridge.coef_)
```

```python
from sklearn.linear_model import Lasso
ridge = Lasso(alpha=0.001).fit(X_train,y_train)
print('训练分数:%f'%(ridge.score(X_train,y_train)))
print('测试分数:%f'%(ridge.score(X_test,y_test)))
print(ridge.coef_)
```

```python
from sklearn import linear_model
import mglearn
X,y = mglearn.datasets.load_extended_boston()
X,y
```

```python
from sklearn.model_selection import train_test_split#将矩阵随机划分为训练子集和测试子集
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1,test_size=0.3)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression#调用sklearn中的线性回归
model = LinearRegression().fit(X_train,y_train)
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))#打分
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=10).fit(X_train,y_train)
print('训练分数:%f'%(ridge.score(X_train,y_train)))
print('测试分数:%f'%(ridge.score(X_test,y_test)))
print(ridge.coef_)
```
