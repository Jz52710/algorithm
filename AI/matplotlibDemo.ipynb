```python
plt.figure(figsize=(4,4))
data = np.arange(10)
plt.plot(data)
```

```python
data = np.arange(10)
plt.plot(data,linewidth=5,color='y',linestyle='--')
plt.legend('y',loc='lower right')
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False
```
